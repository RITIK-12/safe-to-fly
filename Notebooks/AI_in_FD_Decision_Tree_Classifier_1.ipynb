```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('/content/drive/MyDrive/AIFD/Work_2/multiclassData.csv')
df.head()
```

```python
import matplotlib as mpl
font_paths = mpl.font_manager.findSystemFonts()
font_objects = mpl.font_manager.createFontList(font_paths)
font_names = [f.name for f in font_objects]

print(font_names)
```

```python
import seaborn as sns
g = sns.countplot(df['fault'])
g.set_xticklabels(['nominal','d1','d2=0.9','d2=0.8','d2=0.7','d2=0.6','d2=0.5','d2=0.4','d2=0.3'])
plt.rcParams['font.family'] = 'Times'
plt.show()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.40)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2).fit(X_train, y_train)
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```

```python
pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.show()
```
